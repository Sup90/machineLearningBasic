## multiple subplots

**y = x + noise 인 (x, y) 쌍의 toy data 생성**

````python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)

data = np.column_stack((x, y))
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.scatter(x, y, marker='o', c='r', edgecolor='b')
ax1.set_title("Scatter: x vs. y")
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.hist(data, bins=10, label=('x', 'y'))
ax2.set_title("Frequency")
ax2.legend()
ax2.yaxis.tick_right()
````

## meshgrid / contour

입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우에는 3차원 자료가 된다. 3차원 자료를 2차원 평면에 시각화하는 방법은 명암이 아닌 등고선(contour)을 사용하는 방법이다. contour 혹은 contourf 명령을 사용한다. contour는 등고선만 표시하고 contourf는 색깔로 표시한다. 입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 한다.

- meshgrid : coordinate vector 를 이용하여 coordinate matrix 구성

````python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
xx, yy = np.meshgrid(x, y)
Z1 = np.exp(-xx**2 - yy**2)
Z2 = np.exp(-(xx - 1)**2 - (yy - 1)**2)
Z = (Z1 - Z2) * 2
fig, ax = plt.subplots()
CS = ax.contour(xx, yy, Z)
ax.set_title('Simplest countour');
````
